Creating a movie recommendation engine using the iMDB dataset and K Nearest
Neighbor Model. The notebook loads a movie dataset, trains a KNN model on selected features, and recommends the 5 nearest movies to a given movie, based on rating and genre.


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://github.com/ArinB/MSBA-CA-Data/raw/main/CA05/movies_recommendation_data.csv')
df.head()
```

```python
print(df.shape)
print(df.info())
```

```python
#Checking Value counts for Label
df['Label'].value_counts()
```

```python
#Dropping Label Column because serves no purpose
df = df.drop(columns=['Label'])
df.head()
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(n_neighbors=5)

# Select only the numerical features relevant for content-based similarity
features = ['IMDB Rating', 'Biography', 'Drama', 'Thriller', 'Comedy', 'Crime', 'Mystery', 'History']
model.fit(df[features])
```

```python
# Define the custom feature vector based on the user's input
# Make sure the order matches the 'features' list used for training the model
The_Post_features = {
    'IMDB Rating': 7.2,
    'Biography': 1, # Yes
    'Drama': 1,     # Yes
    'Thriller': 0,  # No
    'Comedy': 0,    # No
    'Crime': 0,     # No
    'Mystery': 0,   # No
    'History': 1    # Yes
}

# Convert the custom features dictionary into a numpy array in the correct order
custom_movie_vector = np.array([The_Post_features[feat] for feat in features]).reshape(1, -1)

# Find the k-nearest neighbors using the custom vector
distances, indices = model.kneighbors(custom_movie_vector)

print(f"\nMovies Similar to \"The Post\"")
# Iterate through the recommendations, skipping the first one if it's the movie itself (not applicable here as it's a custom vector)
for i in range(len(distances[0])):
    recommended_movie_index = indices[0][i]
    recommended_movie_name = df.loc[recommended_movie_index, 'Movie Name']
    imdb_rating = df.loc[recommended_movie_index, 'IMDB Rating']
    print(f"{i+1}. {recommended_movie_name} (IMDB Rating: {imdb_rating})")
```
